```python
from src.tsp_utils import TSPUtils
from src.tsp_animator import TSPAnimator
from src.algorithms.genetic import Genetic

def run():
    """
    Run the Genetic Algorithm on the Traveling Salesman Problem (TSP)
    and visualize the optimization process.
    """
    # Parameters
    cities = 20
    population = 20
    generations = 200

    # Initialize problem
    generated = TSPUtils.generate_cities(cities)
    algorithm = Genetic(generated, population, generations)
    history = algorithm.run()

    # Visualize results
    animator = TSPAnimator(generated, history)
    animator.save()


run()
```
